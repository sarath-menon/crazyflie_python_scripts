# CMSIS DSP implementation

## Convert coefficitents

```python
print("No of sections: ", sos.shape[0])
n_sections = sos.shape[0]

# each row of form [b0, b1, b2, a0, a1, a2] corresponds to a 2nd order section 

# Normalize coefficients such that a_0 is 1 and flip the sigs of the a coefficientss
sos_normalized = np.empty(sos.shape)

for (i,section) in enumerate(sos):
    a0 = section[3]
    # normalize 
    sos_normalized[i,:] =  section / a0
    # flip sign of a 
    sos_normalized[i,3:6] = -sos_normalized[i,3:6]

# delete a0 and reshape to single axis array
sos_normalized_a0_removed = np.delete(sos_normalized, 3, 1)
sos_normalized_a0_removed = sos_normalized_a0_removed .reshape(sos_normalized_a0_removed .size)

print("sos: ", sos)
print("sos normalized: ", sos_normalized)
print("sos normalized , a0 removed : ", sos_normalized_a0_removed)

# scale down by 4 
sos_normalized_a0_removed_scaled = sos_normalized_a0_removed/4.0
```

## Cascaded Direct form 1 filter (f32 format) 

```python
from cmsisdsp import arm_biquad_casd_df1_inst_f32, arm_biquad_cascade_df1_init_f32, arm_biquad_cascade_df1_f32

# convert input signal to f32 format
noisy_signal_f32 = gyro_unfiltered[0]
gyro_filtered_cmsis_f32 = np.empty(gyro_unfiltered.shape)

# instance of floating-point Biquad cascade filter
biquadf32 = arm_biquad_casd_df1_inst_f32()

# state array to initialize instance
numStages=n_sections
state=np.zeros(numStages*4, dtype=np.float32)

# convert filter coeff to f32 format
coeff_f32 = sos_normalized_a0_removed.astype(np.float32)

# error checking for coefficient matrix shape
assert coeff_f32.size == 5*numStages, "coeff matrix size incorrect"
# error checking for coefficient matrix shape
assert state.size == 4*numStages, "state size incorrect"

arm_biquad_cascade_df1_init_f32(biquadf32,numStages,coeff_f32,state)

#  apply filter 
gyro_filtered_cmsis_f32[0,:] = arm_biquad_cascade_df1_f32(biquadf32, gyro_unfiltered[0])
gyro_filtered_cmsis_f32[1,:] = arm_biquad_cascade_df1_f32(biquadf32, gyro_unfiltered[1])
gyro_filtered_cmsis_f32[2,:] = arm_biquad_cascade_df1_f32(biquadf32, gyro_unfiltered[2])
```
